## Conducting EDA on the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

I have chosen the 2021 data set because it is the most recent and the year in the data (2019) is not affected by the Covid pandemic.

```python
hf_data_2021 = pd.read_csv('../data/hfi_cc_2021.csv')
print(hf_data_2021.head(30))
```

```python
hf_data_2021.describe()
```

This query in "desired_list" will specify exactly the columsn that I want to work with.  However of note this is not yet casting it into a useable data frame.

```python
desired_list = ['countries', 'ef_rank']

print(hf_data_2021[desired_list].head(30))
```

```python
# Created a new data frame from 2021 data that only includes data from the year 2019.
hf_data_2019 = hf_data_2021[hf_data_2021['year'] == 2019]
```

```python
hf_data_2019.describe()
```

```python
desired_list = ['countries', 'ISO', 'pf_score', 'ef_score']

pf_ef_comparison_2019 = hf_data_2019[desired_list]
```

```python
print(pf_ef_comparison_2019)
```

Attempting to visualize this data with personal freedom as the x and economic freedom the y in a scatterplot.


```python
sns.scatterplot(data=pf_ef_comparison_2019, x= 'pf_score', y= 'ef_score')

plt.xlabel('Personal Freedom Score')
plt.ylabel('Economic Freedom Score')
plt.title('Relationship between Personal and Economic Freedom 2019')
```

```python
plt.figure(figsize=(20, 20))

sns.scatterplot(data=hf_data_2019, x='pf_score', y='ef_score')
plt.xlabel('Personal Freedom Score')
plt.ylabel('Economic Freedom Score')
plt.title('Relationship between Personal Freedom and Economic Freedom (2019)')

# Add country labels as annotations
for i in range(len(hf_data_2019)):
    plt.annotate(hf_data_2019['ISO'][i], (hf_data_2019['pf_score'][i], hf_data_2019['ef_score'][i]))

plt.show()
```
